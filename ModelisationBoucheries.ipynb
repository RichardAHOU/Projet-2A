```python
import pandas as pd
```

```python
Base = pd.read_csv('Projet-2A/boucheries.csv')
Base1 = pd.read_csv('Projet-2A/niveau_de_vie_commune.csv')
niveau = pd.read_csv('Projet-2A/niveau_de_vie_commune.csv', delimiter=';')
convertisseur = pd.read_csv('Projet-2A/correspondance-code-insee-code-postal.csv', delimiter=';')
```

```python
final = convertisseur.loc[convertisseur['Code Département'].isin(['75','92','93','94'])]
final
```

```python
niveau_vie = niveau[['CODGEO','MED14']]
liste_departements=["75", "92", "93", "94"]
department=[]
for ville in niveau_vie.CODGEO:
    department.append(ville[:2])
niveau_vie['departement']=department
niveau_vie=niveau_vie[niveau_vie.departement.isin(liste_departements)]
niveau_vie = niveau_vie.reset_index(drop = True)
ndv = []
for med14 in niveau_vie.MED14:
    ndv.append(float(med14.replace(',','.')))
niveau_vie['MED14']=ndv
niveau_vie   
```

```python
niv = niveau_vie.rename(columns={'CODGEO' : 'Code INSEE'})
niv
```

```python
pref = pd.merge(niv,final,on='Code INSEE')
```

```python
preff = pref[['MED14', 'Code INSEE', 'Code Postal','Population']]
preff
```

```python
boucheries = Base.groupby(["codePostalEtablissement"])["codePostalEtablissement"].count()
boucheries = boucheries.to_frame('Boucheries')
boucheries['Code Postal']=boucheries.index
boucheries = boucheries.rename(columns= {'codePostalEtablissement' : 'Code Postal'})
boucheries['Code Postal'] = boucheries['Code Postal'].astype(str)
boucheries['Boucheries'] = boucheries['Boucheries'].astype(float)
boucheries
```

```python
prefff = pd.merge(preff,boucheries,on='Code Postal')
prefff['MED14'] = prefff['MED14'].astype(float)
prefff['Boucheriesparhabitants']=prefff['Boucheries']/prefff['Population']
prefff
```

```python
import pandas as pd
import statsmodels.formula.api as sm
result = sm.ols(formula="Boucheriesparhabitants ~ MED14", data=prefff).fit()
print(result.params)
```

```python
import seaborn as sns

sns.regplot(x='MED14', y='Boucheriesparhabitants', data=prefff)
```
