```python
def get_location(adress):
    geolocator = Nominatim(user_agent="tutorial")
    location = geolocator.geocode(adress)
    return((location.latitude, location.longitude))

def get_location_all(commerces):
    geom = []
    liste_long=[]
    liste_lat=[]
    for k in range(len(commerces)):
        numero, type_voie, libelle_voie, commune, code_postal = commerces["numeroVoieEtablissement"][k], commerces["typeVoieEtablissement"][k], commerces["libelleVoieEtablissement"][k], commerces["libelleCommuneEtablissement"][k], commerces["codePostalEtablissement"][k]
        try:
            loc1,loc2=get_location((f"{numero} {type_voie} {libelle_voie} {code_postal}"))
        except Exception:
            loc1, loc2 = get_location((f"{code_postal}"))
        geom.append(Point(loc1,loc2))
        liste_long.append(loc2)
        liste_lat.append(loc1)
    commerces = commerces.set_geometry(geom)
    commerces['long']=liste_long
    commerces['lat']=liste_lat
    return(commerces)
```

```python
df = get_location_all(df)
```

```python

for k in range(len(cities)):
    if cities["INSEE_DEP"].iloc[k] == '75':
        cities["MED14"].iloc[k] = niveau_vie.loc[niveau_vie["CODGEO"] == cities["INSEE_ARM"].iloc[k], "MED14"]
    else:
        cities["MED14"].iloc[k] = niveau_vie.loc[niveau_vie["CODGEO"] == cities["INSEE_COM"].iloc[k], "MED14"]
cities = cities.replace(',', '.', regex=True)
cities = cities.loc[cities["MED14"] != ""]
cities = cities.reset_index(drop = True)
```

```python
cities_essential = cities[["NOM", "INSEE_COM", "INSEE_ARM", "geometry", "MED14"]]
```

```python
list_cities = [cities_essential["NOM"].iloc[k] for k in range(len(cities_essential))]
MED_14 = [float(cities_essential["MED14"].iloc[k]) for k in range(len(cities_essential))]
```
